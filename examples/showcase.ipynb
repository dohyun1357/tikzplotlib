# tikzplotlib showcase

[tikzplotlib](https://github.com/dohyun1357/tikzplotlib) converts [matplotlib](https://matplotlib.org/) figures into [PGFPlots](https://www.ctan.org/pkg/pgfplots)/TikZ code that you can drop straight into a LaTeX or ConTeXt document.

Each example below builds a normal matplotlib figure and then prints the TikZ code that `tikzplotlib.get_tikz_code()` produces for it. Use `tikzplotlib.save("figure.tex")` to write the same code to a file.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import tikzplotlib

rng = np.random.default_rng(0)


def show(fig=None):
    """Display the current figure and print its TikZ code."""
    code = tikzplotlib.get_tikz_code()
    plt.show()
    print(code)
```

## Line plot

Markers, line styles, labels, a legend and a title all carry over.

```python
t = np.linspace(0, 2, 40)
plt.plot(t, np.sin(2 * np.pi * t), "o-", label="sin")
plt.plot(t, np.cos(2 * np.pi * t), "s--", label="cos")
plt.xlabel("time (s)")
plt.ylabel("amplitude")
plt.title("Trigonometric functions")
plt.legend()
plt.grid(True)
show()
```

## Scatter plot with a colormap

Per-point colors and sizes become a PGFPlots scatter with a color map.

```python
n = 40
x = rng.normal(size=n)
y = rng.normal(size=n)
colors = rng.uniform(size=n)
sizes = 200 * rng.uniform(size=n)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap="viridis")
plt.title("Scatter")
plt.colorbar(label="value")
show()
```

## Bar chart

```python
labels = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]
plt.bar(labels, values, color="C2")
plt.ylabel("count")
plt.title("Bar chart")
show()
```

## Histogram

```python
data = rng.normal(size=500)
plt.hist(data, bins=20, color="C3", edgecolor="white")
plt.xlabel("value")
plt.ylabel("frequency")
plt.title("Histogram")
show()
```

## Subplots

Multiple axes are emitted as a PGFPlots `groupplot`.

```python
x = np.linspace(0, 2 * np.pi, 50)
fig, axes = plt.subplots(1, 2, figsize=(8, 3))
axes[0].plot(x, np.sin(x), "C0")
axes[0].set_title("sin")
axes[1].plot(x, np.cos(x), "C1")
axes[1].set_title("cos")
show()
```

## Logarithmic axes

```python
x = np.linspace(1, 100, 50)
plt.semilogy(x, x**2, label="$x^2$")
plt.semilogy(x, x**3, label="$x^3$")
plt.xlabel("x")
plt.title("Log-scaled y axis")
plt.legend()
show()
```

## Simplifying dense data with `clean_figure`

`tikzplotlib.clean_figure()` removes points that are outside the axes or visually redundant, which can dramatically shrink the generated code for densely sampled curves.

```python
x = np.linspace(0, 4 * np.pi, 10_000)
plt.plot(x, np.sin(x))
plt.title("10,000 points")

before = tikzplotlib.get_tikz_code()
tikzplotlib.clean_figure()
after = tikzplotlib.get_tikz_code()
plt.show()
print(f"lines of TikZ before clean_figure: {before.count(chr(10)):>6}")
print(f"lines of TikZ after  clean_figure: {after.count(chr(10)):>6}")
```

## LaTeX preamble

Add the packages tikzplotlib relies on to your document preamble. The snippet is available programmatically for both flavors.

```python
print(tikzplotlib.Flavors.latex.preamble())
print("\n--- ConTeXt ---\n")
print(tikzplotlib.Flavors.context.preamble())
```
